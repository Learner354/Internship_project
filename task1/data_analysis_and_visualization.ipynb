```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title('Feature Boxplots for Outlier Detection')
plt.show()
```

```python
from scipy import stats
z_scores = np.abs(stats.zscore(df.select_dtypes(include=['float64', 'int64'])))
outliers = (z_scores > 3).sum()
outliers
```

```python
# Check ranges
print("\nFeature ranges (max - min):")
print(df.max() - df.min())

# Check standard deviations
print("\nFeature standard deviations:")
print(df.std())

# Visual comparison
plt.figure(figsize=(12,6))
df.plot(kind='box', vert=False, logy=True)  # Log scale helps visualize wide ranges
plt.title('Feature Ranges Comparison (Log Scale)')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(15,10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.show()

# Correlation with target
print("\nCorrelation with Class:")
print(corr_matrix['Class'].sort_values(ascending=False))
```

```python
# Time feature analysis
fraud=df[df['Class']==1]
plt.figure(figsize=(12, 6))
sns.distplot(fraud['Time'], bins=100, color='red')
plt.title('Distribution of Transaction Times for Fraud Cases')
plt.xlabel('Time (seconds since first transaction)')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.distplot(fraud['Amount'], bins=50, color='red')
plt.title('Distribution of Transaction Amounts for Fraud Cases')
plt.xlabel('Amount')
plt.ylabel('Density')
plt.show()
```

```python
# Save key statistics to a dictionary
eda_results = {
    'shape': df.shape,
    'missing_values': df.isnull().sum().to_dict(),
    'class_distribution': df['Class'].value_counts().to_dict(),
    'dtypes': df.dtypes.to_dict(),
    'stats': df.describe().to_dict()
}

# You can save this to a JSON file for your report
import json
def convert(o):
    if isinstance(o, np.generic):
        return o.item()
    raise TypeError

# Save to JSON
with open('eda_results.json', 'w') as f:
    json.dump(eda_results, f, default=convert)
```
